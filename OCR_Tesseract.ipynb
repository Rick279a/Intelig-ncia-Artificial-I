```python
import pytesseract
from PIL import Image
import os

# Diretório da pasta com as imagens
diretorio = '/content/Imagens/'

# Lista para armazenar os caminhos das imagens
caminhos_das_imagens = []

# Iterar sobre os arquivos na pasta
for arquivo in os.listdir(diretorio):
    # Verificar se o arquivo é uma imagem (você pode adicionar mais extensões, se necessário)
    if arquivo.endswith('.jpg') or arquivo.endswith('.png'):
        # Construir o caminho completo para a imagem
        caminho_da_imagem = os.path.join(diretorio, arquivo)
        # Adicionar o caminho da imagem à lista
        caminhos_das_imagens.append(caminho_da_imagem)

# Processar as primeiras 20 imagens
for indice, caminho_da_imagem in enumerate(caminhos_das_imagens[:20], start=1):
    # Executar OCR na imagem
    texto = pytesseract.image_to_string(caminho_da_imagem, lang='por')
    # Exibir o texto extraído junto com o número da imagem
    print(f"Imagem {indice}: {texto}")
```
